# tfidf

```python
import numpy as np
import pandas as pd


class LinearSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=200):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None
        
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)
        self.b = 0
        
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.lr * (2 * self.lambda_param * self.w - np.dot(x_i, y[idx]))
                    self.b -= self.lr * y[idx]
                    
    def predict(self, X):
        linear_output = np.dot(X, self.w) - self.b
        return np.sign(linear_output)


# Function to load and preprocess data


# Load and preprocess the datasets
train_data = pd.read_csv("new.tfidf.train.csv", header=None, skiprows=1)
y_train = train_data.iloc[:, 0].values.astype(int)
X_train = train_data.iloc[:, 1:].values.astype(float)



dev_data = pd.read_csv("/home/u1472278/ML/Project/project_data/data/tfidf/tfidf.eval.anon.csv", header=None, skiprows=1)
y_dev = dev_data.iloc[:, 0].values.astype(int)
X_dev = dev_data.iloc[:, 1:].values.astype(float)

#eval_ids = pd.read_csv('/home/u1472278/ML/Project/project_data/data/eval.ids', header=None)

# Separate features and labels
test_data = pd.read_csv("new.tfidf.test.csv", header=None, skiprows=1)
y_test = test_data.iloc[:, 0].values.astype(int)
X_test = test_data.iloc[:, 1:].values.astype(float)

# Train the SVM
svm = LinearSVM()
svm.fit(X_train, y_train)

# Test accuracy
predictions_test = svm.predict(X_test)
test_accuracy = np.mean(predictions_test == y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Predict on evaluation dataset
predictions_eval = svm.predict(X_dev)

# Concatenate eval IDs with predictions
predicted_labels_int = predictions_eval.astype(int)

predicted_labels_df = pd.DataFrame(predicted_labels_int, columns=['label'])
eval_ids = pd.read_csv("data/eval.ids", header=None, names=['example_id'])

submission_df = pd.concat([eval_ids['example_id'], predicted_labels_df], axis=1)

# Save the concatenated DataFrame to a CSV file
submission_df.to_csv('submission_svm_tfidf.csv', index=False)
```
